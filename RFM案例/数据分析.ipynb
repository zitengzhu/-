```python
# 将销售金额小于1的数据设置为nan
# 订单金额列中 ≤1 的值被替换为 NaN
y_2015['订单金额']=y_2015['订单金额'][y_2015['订单金额']>1]  # 将订单金额小于一的数据设置为nan
y_2015.dropna(inplace=True)
y_2015.info()
y_2016['订单金额']=y_2016['订单金额'][y_2016['订单金额']>1]  # 将订单金额小于一的数据设置为nan
y_2016.dropna(inplace=True)
y_2016.info()
y_2017['订单金额']=y_2017['订单金额'][y_2017['订单金额']>1]  # 将订单金额小于一的数据设置为nan
y_2017.dropna(inplace=True)
y_2017.info()
y_2018['订单金额']=y_2018['订单金额'][y_2018['订单金额']>1]  # 将订单金额小于一的数据设置为nan
y_2018.dropna(inplace=True)
y_2018.info()

# 索引不连续造成了 30574 entries, 0 to 30773 这种现象
```

```python
# 添加参考日期作为计算购买间隔的标准(以店铺开业的最后一天作为购买一起的最大值 每年的最后一天设置一个参考点)
# 只能在合并表之前进行操作 不然回导致 参考时间设置 可能导致出现 全是 最新日期的情况 实例: 15年的提交日期 参考时间 为 18年开业的最后一天
y_2015['year_data_max']=y_2015['提交日期'].max()
y_2016['year_data_max']=y_2016['提交日期'].max()
y_2017['year_data_max']=y_2017['提交日期'].max()
y_2018['year_data_max']=y_2018['提交日期'].max()
```

```python
# 为了后续操作数据的方便将四个表进行合并
df_concat=pd.concat([y_2015,y_2016,y_2017,y_2018],ignore_index=True)   # 合并表单之后需要重置索引
df_concat
```

```python
# 为了号区分 给新合并的表格添加年份信息 标注 数据的记录年份
df_concat['year']=df_concat['提交日期'].dt.year   # .dt.year 可以直接从 datatime类型的数据中获取年份信息
df_concat
```

```python
# 根据参考日期获取出用户距离上次购买的间隔
df_concat['time_interval']=df_concat['year_data_max']-df_concat['提交日期']
df_concat
# 为了后期操作相变将time_interval改为整数形式
df_concat['time_interval'] = df_concat['time_interval'].dt.days
```

```python
# 根据 会员ID 和 year 来进行分类 获取 每年 的消费金额,购买次数和购买的最小间隔 以此来进行评分  即 RFM
# RFM   recency(最近一个的购买时间)   frequency(购买次数) monetary(购买金额)
rfm_gb = df_concat.groupby(['year','会员ID'],as_index=False).agg({
    'time_interval':'min',
     '订单号':'count',
    '订单金额':'sum'
})
rfm_gb
```

```python
# 修改列名
rfm_gb.columns=['year','会员ID','r','f','m']
rfm_gb
```

```python
# 查看rfm的数据
rfm_gb.iloc[:,2:].describe().T
# T 将行和列互换便于观察数据
```

```python
# 数据查看完毕开始划分区间 并 给出frm评分 
# r 最近一次购买的时间 值越小分越高
# f 购买次数 值越大分越高
# m 购买金额 值越大分越高
```

```python
# 思路一:系统自动划分区间 (了解)
# 使用cut()函数可以让系统自动划分区间 参数一是 要划分的区间   参数二 bins是划分的区间个数
pd.cut(rfm_gb['r'],bins=3)
```

```python
# 思路二: 自己设置区间 包右不包左
r_bins=[-1,79,255,365]
f_bins=[0,2,5,130]
m_bins=[0,69,1199,206252]
# 给 r 划分区间并打分   值越小分越高
rfm_gb['r_label']=pd.cut(rfm_gb['r'],bins=r_bins,labels=[i for i in range(len(r_bins)-1,0,-1)])
# 给 f 划分区间并打分   值越大分越高
rfm_gb['f_label']=pd.cut(rfm_gb['f'],bins=f_bins,labels=[i for i in range(1,len(f_bins))])
# 给 m 划分区间并打分   值越大分越高
rfm_gb['m_label']=pd.cut(rfm_gb['m'],bins=m_bins,labels=[i for i in range(1,len(m_bins))])
rfm_gb
```

```python
# 计算rfm的值
# 因为数据为int类型不能拼接先要转化为str类型
rfm_gb['r_label']=rfm_gb['r_label'].astype(str)
rfm_gb['f_label']=rfm_gb['f_label'].astype(str)
rfm_gb['m_label']=rfm_gb['m_label'].astype(str)
rfm_gb['rfm_group']=rfm_gb['r_label']+rfm_gb['f_label']+rfm_gb['m_label']
rfm_gb
```

```python
# 保存数据到excel文件中
rfm_gb.to_excel('rfm分析结果.xlsx',index=False)
```

```python
# 进行数据可视化
data = rfm_gb.groupby(['year','rfm_group'],as_index=False).agg({'会员ID':'count'})
data.columns=['year','rfm.group','number']
data['number']=data['number'].astype(int)
```

```python
# 进行绘制
# 5.3 绘制图形
# 颜色池
range_color = ['#313695', '#4575b4', '#74add1', '#abd9e9', '#e0f3f8', '#ffffbf',
               '#fee090', '#fdae61', '#f46d43', '#d73027', '#a50026']

range_max = int(data['number'].max())
c = (
    Bar3D()#设置了一个3D柱形图对象
    .add(
        "",#图例
        [d.tolist() for d in data.values],#数据
        xaxis3d_opts=opts.Axis3DOpts(type_="category", name='分组名称'),#x轴数据类型，名称，rfm_group
        yaxis3d_opts=opts.Axis3DOpts(type_="category", name='年份'),#y轴数据类型，名称，year
        zaxis3d_opts=opts.Axis3DOpts(type_="value", name='会员数量'),#z轴数据类型，名称，number
    )
    .set_global_opts( # 全局设置
        visualmap_opts=opts.VisualMapOpts(max_=range_max, range_color=range_color), #设置颜色，及不同取值对应的颜色
        title_opts=opts.TitleOpts(title="RFM分组结果"),#设置标题
    )
)
c.render() 		      # 数据保存到本地的网页中.
# c.render_notebook() #在notebook中显示
```
