```python
import pandas as pd
```

```python
# 拿到数据的第一步就先接收数据然后检查数据
```

```python
# 读取excel表格并获取数据
shop = pd.read_excel('shop.xlsx',None)
shop
```

```python
# 获取数据之后发现数据有3个sheet表格 分别为 订单信息 货物信息 顾客信息
# 开始定义变量接收信息
```

```python
# 定义变量接收订单信息  
orders = shop['订单信息']
orders
```

```python
# 定义变量接收 货物信息
goods = shop['货物信息']
goods
```

```python
# 定义变量接收顾客信息
cus = shop['顾客信息']
cus
```

```python
# 数据接收完毕开始检查数据是否正确
```

```python
# 检查订单信息的数据
orders.info()
```

```python
# 检查货物信息
goods.info()
```

```python
# 检查顾客信息
cus.info()
```

```python
# 检查之后发现订单数应该为int数据但数据的结果为object数据说明
# 顾客信息中的订单数这一列数据有错误
```

```python
# 因为顾客信息中的订单数这一列数据有错误现在开始编写代码定位有问题的数据
# 因sheet表格内的数据过多可以通过循环查找数据
# index 代表没一行的索引 row 代表每一行的数据  cus.iterrows() 代表循环的是cus中的数据
for index,row in cus.iterrows():
    if type(row['订单数']) != int:
        print('这一行数据有问题',index,row)
```

```python
# 查看到有错位的数据
# 第46行的订单数出现错误 
# 查看到数据错误后开始修改
# 要找到目标数据的全部订单数  找数据显示为1的订单  在订单信息一个sheet表格内查找
```

```python
orders[(orders.客户ID==6243) & (orders.订单状态==1)].shape
```

```python
# 使用loc方法查找数据
cus.loc[46,"订单数"]=orders[(orders.客户ID==6243) & (orders.订单状态==1)].shape[0]
```

```python
cus.loc[46,'订单数']
```

```python
# 数据修改完毕之后开始生成新的列表
```

```python
# 将修改好的excel表格写入
```

```python
# 创建新的excel表格
writer = pd.ExcelWriter("shop_new.xlsx")
# excel表格创建完毕之后开始写入数据
orders.to_excel(
    writer,
    sheet_name="订单信息",
    index = False
)
goods.to_excel(
    writer,
    sheet_name="货物信息",
    index = False
)
cus.to_excel(
    writer,
    sheet_name="顾客信息",
    index = False
)
```

```python
# 保存文件
writer._save()
```
