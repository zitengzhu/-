```python
# 导入pandas模块
import pandas as pd
```

```python
# 经过之前两次的查看已经就数据全部修改完毕
# 现在读取excel表格开始 
shop = pd.read_excel('shop_new_1.xlsx',None)
shop
```

```python
# 数据读取完毕开始接受数据
```

```python
# 订单信息
orders = shop['订单信息']
orders
```

```python
# 货物信息
goods = shop['货物信息']
goods
```

```python
# 顾客信息
cus = shop['顾客信息']
cus
```

```python
# 下面完成甲方需求
# 分析下平台的订单情况、商品情况
# 获取打折商品和不打折商品的销量前十和后十个
# .groupby("货物名称") 的意思为按货物名称分组 相同货物名称的货物为一组
# .size()可以获取 分组之后每一组的数量
# .sort_values() 可以进行排序 默认是升序
```

```python
# 获取不打折商品销量的后10个
goods[goods.优惠额度=="0"].groupby('货物名称').size().sort_values().head(10)
```

```python
# 获取不打折商品销量的前10个
goods[goods.优惠额度=="0"].groupby('货物名称').size().sort_values().tail(10)
```

```python
# 获取打折商品销量的后十个
goods[goods.优惠额度!="0"].groupby('货物名称').size().sort_values().head(10)
```

```python
# 获取打折商品销量的前十个
goods[goods.优惠额度!="0"].groupby('货物名称').size().sort_values().tail(10)
```

```python
总结：
    销量前十名的商品是 xxx  可以考虑继续推广  增加商品的铺货
    销量后十名的商品  xxx   可以考虑打折  也可以考虑下架   如果对这个商品有信心 也可以继续推广
```

```python
# 核心需求：分离在本电商平台购物的无价值用用户
# 无价值用户标准 在成交的所有商品中 优惠商品的占比在75%及以上的
```

```python
# 想要查看购买的优惠商品和无优惠商品的数量 先要获取全部成交商品的数据
res = orders[orders.订单状态==1]
res
```

```python
# 获取到数据后要使用数据交叉表更直观的看出客户购买货物的情况
all = pd.crosstab(res.客户ID,res.优惠类型)
all # 0和1表示的是优惠类型  0是无优惠 1是有优惠
```

```python
# 表格制作完成开始获取有价值用户的ID
# all[0] 无优惠订单的数量  all[1]有优惠订单的数量     all[0]表示获取优惠类型为0的数据
```

```python
p = all[all[0]/(all[0]+all[1])>=0.75] 
p
```

```python
# 发现不好统计客户ID加入index索引
p = all[all[0]/(all[0]+all[1])>=0.75].index
p
```

```python
# 加入values获取index的值
p = all[all[0]/(all[0]+all[1])>=0.75].index.values
p
```

```python
# 现在数据方便统计了 获取到了所有有价值用户的ID
# 加入一列新数据标记哪些是有价值用户哪些是无价值用户
# 添加新的列数据
cus['客户分类'] = 0   # 在顾客信息这个sheet表格内添加一列 客户分类的数据 数据的值是0
cus
```

```python
# 将有价值用户和无价值用户的对应数据导入表格 重写客户分类的数据
# 因为甲方主要需求为分离用户  所以要重置索引 
# 数字索引重置为 用客户ID进行索引 用来查看是有价值用户和无价值用户
# 重置索引要使用set_index
```

```python
cus2 = cus.set_index('客户ID')
cus2
```

```python
# 索引设置完毕开始进行 雷夫分类一列数据的重写
cus2.loc[p,"客户分类"]=1
cus2
```

```python
# 查看有价值用户和无价值用户的具体个数
data = cus2.groupby('客户分类')
data
```

```python
# 加入.mean()可以获取具体数据
data = cus2.groupby('客户分类').mean()
data
```

```python
#.rename(0:'无价值用户',1:'有价值用户') 对0和1进行重命名  .astype(int)将数据转化为整形
data = cus2.groupby('客户分类').mean().rename({0:'无价值用户',1:'有价值用户'}).astype(int)
data
```

```python
# 甲方需求解决完毕开始生成excel表格
w=pd.ExcelWriter("shpo_end.xlsx")
orders.to_excel(w,sheet_name="订单信息",index=False)
goods.to_excel(w,sheet_name="货物信息",index=False)
cus2.to_excel(w,sheet_name="客户信息",index=False)
w._save()
```
