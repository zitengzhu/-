```python
# 绘制图形时候必须要的模块  没有的话图形显示不出来
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 要想利用jupyter绘制图形先要导入绘制图形所需要的模块 
#  import 后面加入的星号 代表导入全部绘图模块
from pyecharts.charts import *
```

```python
# 导入自定义图形外观的模块
from pyecharts import options as opts
```

```python
# 绘制柱状图
# 1.明确数据(你的 行数据 和 列数据 分别是什么)
# 自定义数据
# 定义行数据 也就是x轴数据
x_data = ['apple','huawei','xiaomi','oppo','vivo','meizu']
# 定义列数据 也就是y轴数据
y_data = [153,110,516,666,12,71]
# 数据全部定义完成之后 开始选择想要绘制的表格图形
# 绘制柱桩图  Bar是柱状图
chart = Bar()
# 图形绘制完毕之后开始把定义好的数据添加到表格内
# 添加x轴的表格数据
chart.add_xaxis(x_data)
# 添加y轴的表格数据  变量前面 一定要加上一个字符串 和x轴的添加数据做区分
chart.add_yaxis('',y_data)
```

```python
# 所有数据准备完毕调用底层代码生成表格
chart.load_javascript()
```

```python
# # 所有数据准备完毕调用底层代码生成表格
chart.render_notebook()
```

```python
# 绘制柱状图
# 1.明确数据(你的 行数据 和 列数据 分别是什么)
# 自定义数据
# 定义行数据 也就是x轴数据
x_data = ['apple','huawei','xiaomi','oppo','vivo','meizu']
# 定义列数据 也就是y轴数据                             
y_data = [opts.BarItem( name = 'apple',value = 153, itemstyle_opts= opts.ItemStyleOpts(color='red'),label_opts = opts.LabelOpts(font_size=30)),
          opts.BarItem(name = 'huawei',value=110),
          opts.BarItem(name = 'xiaomi',value=516),
          opts.BarItem(name = 'oppo',value=666),
          opts.BarItem(name = 'vivo',value=12),
          opts.BarItem(name = 'meizu',value=71)
         ]
# 数据全部定义完成之后 开始选择想要绘制的表格图形
# 绘制柱桩图  Bar是柱状图
chart = Bar()
# 图形绘制完毕之后开始把定义好的数据添加到表格内
# 添加x轴的表格数据
chart.add_xaxis(x_data)
# 添加y轴的表格数据  变量前面 一定要加上一个字符串 和x轴的添加数据做区分
chart.add_yaxis('',y_data)
```

```python
chart.load_javascript()
```

```python
chart.render_notebook()
```

```python
# 除了生成 单个柱状图外 还可以在一个x轴的元素内生成左右两个柱状图
```

```python


# 定义x轴数据
x_data = ["Apple","Huawei","Xiaomi","Oppo","Vivo","Meizu"]
# 定义左右两个不同的y轴数据
# 定义y轴左边的数据
y_data_left = [123,153,89,107,98,23]
# 定义y轴右边的数据
y_data_right = [13,15,8,10,9,2]
# 选择图形
chart = Bar()
# 数据全部准备完毕开始添加数据
chart.add_xaxis(x_data)
# 添加左侧数据    位置 要添加的数据  索引为0在最左侧
chart.add_yaxis('左侧',y_data_left,yaxis_index=0)
# 添加右侧数据    位置   要添加的数据  索引为1在最右侧
chart.add_yaxis('右侧',y_data_right,yaxis_index=1)
# 添加坐标轴   自动帮你划分y轴数据的区间
chart.extend_axis(yaxis=opts.AxisOpts())
```

```python
chart.load_javascript()
chart.render_notebook()
```

```python
#绘制折线图
## 自定义行数据
x_data=["Apple","Huawei","Xiaomi","Oppo","Vivo","Meizu"]
# 自定义列数据
y_data=[123,153,89,107,98,23]
# 2. 选择要绘制的图形   Line是折线图图
chart=Line()
# 开始绘图  把行数据 列数据 添加到 图形中   (chart变量中)
# 添加x轴的数据
chart.add_xaxis(x_data)
# 添加y轴的数据    变量前面 一定要加上一个字符串
chart.add_yaxis("",y_data)
```

```python
chart.load_javascript()
chart.render_notebook()
```

```python
# 带有阴影的折线图
# 自定义行数据
x_data=["Apple","Huawei","Xiaomi","Oppo","Vivo","Meizu"]
# 自定义列数据
y_data=[123,153,89,107,98,23]
# 2. 选择要绘制的图形   Line是折线图图
chart=Line()
# 开始绘图  把行数据 列数据 添加到 图形中   (chart变量中)
# 添加x轴的数据
chart.add_xaxis(x_data)
# 添加y轴的数据    变量前面 一定要加上一个字符串        括号里面的参数就是 图形阴影情况
chart.add_yaxis("",y_data, areastyle_opts=opts.AreaStyleOpts(opacity=0.5))   #数字越小 阴影越浅  数字越大 阴影越深
```

```python
chart.load_javascript()
```

```python
chart.render_notebook()
```

```python
import numpy as np
import pandas as pd  
import matplotlib.pyplot as pit # 绘制饼状图的模块
```

```python
age_new = np.random.normal(25,5,100).astype(int)
score = np.random.normal(20,8,100).astype(int)
data = pd.DataFrame({"age":age_new,"score":score})
data2 = data.copy()
data2['age_cut']=pd.cut(data2.age,bins=[1,12,18,36,50])
print(data2)
```

```python
# 绘制饼状图
data2.age_cut.value_counts().plot.pie(
    figsize=(4,3),   #可以控制图形的大小
    autopct="%1.3f%%" #可以控制文字的百分比显示
)
```

```python
# 绘制柱状图 
# 根据年龄进行 统计数量
data2.groupby('age_cut').count()
```

```python
# 根据年龄中的成绩 进行统计数量
data2.groupby('age_cut')['score'].count()
```

```python
# 加入['score']后年龄的一列数据消失了成绩一列的数据显示
```

```python
# 通过分组之后的数据进行绘图
data2.groupby('age_cut')['score'].count().plot.bar(
    figsize=(3,2)
)
```
