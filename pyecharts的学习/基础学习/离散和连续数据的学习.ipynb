```python
import numpy as np
import pandas as pd  
```

```python
# 生成年龄的数据
# random是随机模块 normal是正常值 
# 25 代表生成的数据的在25左右
# 5 代表生成的数据的值和25的差是5左右
# 100代表生成100个这样的值
age = np.random.normal(25,5,100)
age
```

```python
# 生成年龄的数据
# random是随机模块 normal是正常值 
# 25 代表生成的数据的在25左右
# 5 代表生成的数据的值和25的差是5左右
# 100代表生成100个这样的值
age = np.random.normal(25,5,100)
# 根据上一行打印数来的代码可知数据类型为浮点型 的小数 不方便观察
# 可以使用 .astype()将数据变为整数 括号内是想要转化的数据类型
age_new = np.random.normal(25,5,100).astype(int)
age_new
```

```python
# 生成成绩的数据
score = np.random.normal(20,8,100).astype(int)
score
```

```python
# 成绩和年龄数据生成完毕将他们转化为二维列表
data = pd.DataFrame({"age":age_new,"score":score})  # 将二维列表是字典数据所以要加上{}
print(data)
```

```python
# 使用.copy()方法可以粘贴数据去做数据处理
```

```python
data_new=data.copy() # 赋值data_new的数据
data_new
```

```python
# .cut() 可以自定义区间进行数据的分类
```

```python
# 对数据年龄这一列数据按照一定区间值进行分类   （.cut() 可以自定义区间进行数据的分类）
#  data_new.age 代表data_new这一个二维数据中age这一列数据  
# bins=[]是进行自主指定区间值进行区间设置时使用的方法
#  [1,12,18,36,50] 代表 设置的区间分别为 1,12 12,18 18,36 36,50  左边为开区间 后边为闭区间
pd.cut(data_new.age,bins=[1,12,18,36,50])
```

```python
# 数据进行区间的分类之后可以把新生成的列区间，添加到列表中  
# data_new['区间'] 中中括号的区间 代表在data_new这个二维数据中添加了一列名为 区间 的列数据
data_new['区间']=pd.cut(data_new.age,bins=[1,12,18,36,50])
data_new
```

```python
# 调用.cat.categories可以查看设置的分类区间的具体数量
data_new.区间.cat.categories
# 由数据可知设置了四个区间
```

```python
# .value_counts()可以统计区间中数据的个数
data_new.区间.value_counts()
```

```python
data2 = data.copy()
print(data2)
```

```python
# pd.qcut 也可以进行数据的区间设置 但与cut()不同的是 
# .qcut不可以指定区间值(区间值由系统生成)，但可以指定有几个区间
```

```python
# data2是要进行数据区间设置的值   4代表设置4个区间
pd.qcut(data2.age,4)
```

```python
# 使用labels可以给区间起名字
pd.qcut(data2.age,4,labels=['少年','青年','中年','老年'])
```

```python
# 可以将生成的区间加入到二维数据中
data2['年龄']=pd.qcut(data2.age,4,labels=['少年','青年','中年','老年'])
data2
```

```python
# 查看代码中各区间中包含的数据个数
data2.年龄.value_counts()
```

```python
# 总结 
# cut 可以自主设置区间值进行数据区间的设置
# qcut不可以自主设置区间值进行数据区间的分类 但可以设置区间的个数
```
